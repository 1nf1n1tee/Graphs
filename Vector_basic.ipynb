#**Vectors**



```python
import numpy as np
from matplotlib import pyplot as plt

def dot(a1,a2):
  ans = np.sum(a1*a2)
  return ans
def lateral(a1,a2):
  if dot(a1,a2) == 0:
    return True
  else:
    return False
def cross(a1,a2):
  ans = np.cross(a1,a2)
  return ans
def vadd(a1,a2):
  return a1+a2
def vsub(a1,a2):
  return a1-a2
```

```python
arr1 = [np.random.uniform(-10000,10000,3) for i in range(1000)]
arr2 = [np.random.uniform(-10000,10000,3) for i in range(1000)]
ans = [dot(arr1[i],arr2[i]) for i in range(100)]
```

```python
plt.hist(ans,bins=100,edgecolor='black')
plt.show()
```
