```python
import numpy as np
import matplotlib.pyplot as plt

def Monte_carlo(n,x,a=0,b=np.pi):
  y = np.sin(x)
  v = np.sum(y)*(b-a)/n
  return v
arr = []
```

```python
a = 0
b = np.pi
n = 1000
def area_gen(n):
  for i in range(n):
    val = Monte_carlo(n,np.random.uniform(a,b,n))
    arr.append(val)
inp = 10
#inp = int(input("Enter a range: "))
for i in range(inp):
  area_gen(n)
```

```python
plt.hist(arr,bins=50,edgecolor = "black")
plt.show()
print(len(arr))
```
